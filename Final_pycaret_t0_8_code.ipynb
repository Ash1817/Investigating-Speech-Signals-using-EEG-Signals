```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Speech_data.csv")
df.shape
```

```python
from pandas import read_csv
from pycaret.classification import *
from sklearn.ensemble import ExtraTreesClassifier

data = df.sample(frac=0.9, random_state=123)
data_unseen = df.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))

numericals = df.select_dtypes(include=['number']).columns.tolist()

exp1 = ClassificationExperiment()
exp1.setup(data, target='Output', session_id=555, train_size=0.8, preprocess=False, numeric_features=numericals)

top3_exp1 = exp1.compare_models()
```
